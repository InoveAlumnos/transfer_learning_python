# Demo de Yolo (you only looks once)

Detección de objetos en imagenes<br>

Esta demo está basada en el siguiente repositorio<br>
[GitHub -Oficial] (https://github.com/ultralytics/ultralytics/blob/main/README.md)

[DOCU-OFICIAL] (https://docs.ultralytics.com/es)
<br>

YOLO: Es un acrónimo de "You Only Look Once", que es un enfoque de detección de objetos en visión por computadora. YOLO divide la imagen en una cuadrícula y realiza predicciones de objetos en cada celda de la cuadrícula.

# Importar las herramientas

```python
# Se carga el modelo YOLO con los pesos preentrenados utilizando Ultralytics.
# Se realiza la predicción en el video de entrada utilizando el modelo.
# Se guarda el video resultante en una ubicación específica.
# Se utiliza cv2_imshow para mostrar el video en la celda de salida de Colab.

from ultralytics import YOLO
import yaml
import cv2
from google.colab.patches import cv2_imshow
```

# Se crea una instancia del modelo YOLOv8

```python
# Se crea una instancia del modelo YOLOv8 utilizando la clase YOLO.
# El argumento "yolov8n.pt" especifica el archivo de pesos preentrenados
# que se utilizarán para inicializar el modelo. Esto significa que el modelo
# ya ha sido entrenado en un conjunto de datos previo y ahora está listo para realizar predicciones.
model = YOLO("yolov8n.pt")
```

# Descargar la imagen de prueba

```python
import os
import platform

if os.access('street.jpg', os.F_OK) is False:
    if platform.system() == 'Windows':
        !curl https://raw.githubusercontent.com/InoveAlumnos/transfer_learning_python/main/imagenes/street.jpg > street.jpg
    else:
        !wget street.jpg https://raw.githubusercontent.com/InoveAlumnos/transfer_learning_python/main/imagenes/street.jpg
```

```python
# La función predict de un modelo YOLOv8 en Ultralytics se utiliza para realizar predicciones en una imagen.
# street.jpg: Especifica la ruta de la imagen en la que deseas realizar predicciones.
# save=True: Este parámetro indica que deseas guardar el resultado visual de la predicción.
# La imagen resultante se guarda en la carpeta runs/detect/exp
# save_txt=True: Este parámetro indica que deseas guardar la información de las predicciones en un archivo de texto.
# Este archivo contendrá información sobre las coordenadas de las cajas delimitadoras y las clases detectadas.
# El archivo de texto se guarda en la misma carpeta runs/detect/exp.
model.predict("street.jpg", save=True, save_txt=True)
```

```python
# Ruta de la imagen
image_path = "/content/runs/detect/predict/street.jpg"

# Cargar la imagen con OpenCV
image = cv2.imread(image_path)

# Mostrar la imagen
cv2_imshow(image)

# Estas líneas son específicas de OpenCV. cv2.waitKey(0) espera a que se presione cualquier tecla (0 indica que espera indefinidamente)
# y cv2.destroyAllWindows() cierra todas las ventanas de visualización de OpenCV.
# En este caso, permiten mantener la ventana de visualización abierta hasta que se presione una tecla y luego cerrarla.
cv2.waitKey(0)
cv2.destroyAllWindows()
```
