# Demo de Yolo (you only looks once)

Detección de objetos en imagenes<br>

Esta demo está basada en el siguiente repositorio<br>
[GitHub -Oficial] (https://github.com/ultralytics/ultralytics/blob/main/README.md)

YOLO: Es un acrónimo de "You Only Look Once", que es un enfoque de detección de objetos en visión por computadora. YOLO divide la imagen en una cuadrícula y realiza predicciones de objetos en cada celda de la cuadrícula.

# Comando de la interfaz de línea de comandos de Ultralytics y YOLO para realizar detección de objetos en un archivo de video.

```python
# !yolo: Indica que estás ejecutando un comando en la interfaz de línea
# de comandos de Ultralytics y YOLO. La exclamación (!) es a menudo utilizada
# en entornos como Jupyter Notebooks o Google Colab para ejecutar comandos de
# la línea de comandos desde el entorno de la celda.

# detect: Especifica que estás realizando el proceso de detección de objetos.

# predict: Indica que estás realizando una predicción utilizando el modelo.

# model=yolov8m.pt: Especifica el modelo a utilizar durante la predicción.
# En este caso, el modelo se carga desde el archivo de pesos "yolov8m.pt".
# Este archivo contiene los pesos preentrenados del modelo YOLOv8.

# source="/content/video.mp4": Indica la fuente de entrada para la predicción.
# En este caso, la fuente es un archivo de video ubicado en "/content/video.mp4".
# El modelo realizará detección de objetos en cada fotograma del video.

!yolo detect predict model=yolov8m.pt source="/content/video.mp4"
```
