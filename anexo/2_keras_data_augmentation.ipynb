```python
import os
import platform

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from keras.preprocessing.image import ImageDataGenerator
```

```python
# Mostrar 4 imagenes afectadas por data augmentation
def plot_imagenes(data_generator, images):
    data_generator.fit(images)
    image_iterator = data_generator.flow(images)
    
    fig = plt.figure(figsize=(18,18))    
    for i in range(4):
        ax = fig.add_subplot(1, 4, i+1)
        ax.imshow(image_iterator.next()[0].astype('int'))
        ax.axis('off')
    plt.show()
```

```python
if os.access("inove_icono.jpg", os.F_OK) == False:
    if platform.system() == 'Windows':
        !curl https://raw.githubusercontent.com/InoveAlumnos/transfer_learning_python/main/inove_icono.jpg > inove_icono.jpg
    else:
        !wget inove_icono.jpg https://raw.githubusercontent.com/InoveAlumnos/transfer_learning_python/main/inove_icono.jpg
    
image = mpimg.imread("inove_icono.jpg")

# Creating a dataset which contains just one image.
imagenes = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

plt.imshow(imagenes[0])
plt.show()
```

```python
# Rotacion
data_generator = ImageDataGenerator(rotation_range=90)
plot_imagenes(data_generator, imagenes)
```

```python
# √Ångulo de corte
data_generator = ImageDataGenerator(shear_range=45)
plot_imagenes(data_generator, imagenes)
```

```python
# Desplazamiento horitonzal
data_generator = ImageDataGenerator(width_shift_range=0.5)
plot_imagenes(data_generator, imagenes)
```

```python
# Desplazamiento vertical
data_generator = ImageDataGenerator(height_shift_range=0.5)
plot_imagenes(data_generator, imagenes)
```

```python
# Zoom
data_generator = ImageDataGenerator(zoom_range=0.5)
plot_imagenes(data_generator, imagenes)
```

```python
# Espejado horizontal
data_generator = ImageDataGenerator(horizontal_flip=True)
plot_imagenes(data_generator, imagenes)
```
